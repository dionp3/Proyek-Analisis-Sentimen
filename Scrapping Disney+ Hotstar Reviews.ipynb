Kode berikut digunakan untuk mengambil data terbaru sebanyak 20.000 ulasan dengan bahasa dan lokasi Indonesia:

```python
from google_play_scraper import reviews, Sort
app_ID = 'in.startv.hotstar.dplus'

result, _ = reviews(
    app_ID,
    lang='id',
    country='id',
    sort = Sort.NEWEST,
    count= 20000
)

print(f"Jumlah data yang diambil: {len(result)}")
```

üßπ Penyaringan Data
Data disimpan dalam DataFrame dan disaring berdasarkan tahun:

```python
import pandas as pd
df_reviews = pd.DataFrame(result)
df_2024_2025 = df_reviews[(df_reviews['at'].dt.year == 2024) | (df_reviews['at'].dt.year == 2025)]
df_reviews.head()
```

üíæ Penyimpanan Data
Data kemudian disimpan dalam file .csv:

```python
df_reviews.to_csv('reviews_disneyhotstar.csv', index =False)

from google.colab import files
files.download('reviews_disneyhotstar.csv')
```
